```python
df1['Case_Fatality_Ratio']=df1['Case_Fatality_Ratio'].fillna(df1['Case_Fatality_Ratio'].mean())
df1['Incident_Rate']=df1['Incident_Rate'].fillna(df1['Incident_Rate'].mean())
df1['Lat']=df1['Lat'].fillna(df1['Lat'].mean())
df1['Long_']=df1['Long_'].fillna(df1['Long_'].median())
df1['FIPS']=df1['FIPS'].fillna(df1['FIPS'].mean())
df1.drop(['Province_State'],axis=1,inplace=True)
df1
```

```python
df1.dropna(inplace=True)
df1
```

```python
df1.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
cc=['Country_Region','Combined_Key']
le=LabelEncoder()
for column in cc:
    df1[column]=le.fit_transform(df1[column])
```

```python
df1.reset_index(drop=True,inplace=True)
df1
```
